```python
import pandas as pd
df = pd.read_excel('dados_star_schema_pratica.xlsx')
df['data'] = pd.to_datetime(df['data'])
df
```

```python
dim_cliente = df[['cliente','cidade']].drop_duplicates().reset_index(drop=True)

#chave surrogate

dim_cliente['cliente_id'] = dim_cliente.index + 1
```

```python
dim_produto = df[['produto','categoria']].drop_duplicates().reset_index(drop=True)

dim_produto['produto_id'] = dim_produto.index + 1
```

```python
dim_data = df[['data']].drop_duplicates().reset_index(drop=True)

dim_data['data'] = pd.to_datetime(dim_data['data'])

dim_data['ano'] = dim_data['data'].dt.year
dim_data['mes'] = dim_data['data'].dt.month
dim_data['dia'] = dim_data['data'].dt.day

dim_data['data_id'] = dim_data.index + 1
```

```python
#Tabela fato

fato = df.merge(dim_cliente, on=['cliente','cidade'])\
         .merge(dim_produto, on=['produto','categoria'])\
         .merge(dim_data, on=['data'])

fato = fato[['pedido_id', 'cliente_id', 'produto_id', 'data_id', 'valor']]

fato
```
